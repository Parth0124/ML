```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
### make dataset

from sklearn.datasets import make_classification
```

```python
x,y=make_classification(n_samples=1000, n_features=2, n_classes=2, n_clusters_per_class=1, n_redundant=0)
```

```python
x
```

```python
y
```

```python
sns.scatterplot(x=pd.DataFrame(x)[0], y=pd.DataFrame(x)[1], hue=y)
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC(kernel='linear')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=15)
```

```python
svc.fit(x_train,y_train)
y_predicted=svc.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
classification_report=classification_report(y_test,y_predicted)
confusion_matrix=confusion_matrix(y_test,y_predicted)
print(classification_report)
print(confusion_matrix)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
polynomial=SVC(kernel='poly')
polynomial.fit(x_train,y_train)
## Prediction
y_predicted2=polynomial.predict(x_test)
print(classification_report(y_test,y_predicted2))
print(confusion_matrix(y_test,y_predicted2))
```

```python
rbf=SVC(kernel='rbf')
rbf.fit(x_train,y_train)
## Prediction
y_predicted3=rbf.predict(x_test)
print(classification_report(y_test,y_predicted3))
print(confusion_matrix(y_test,y_predicted3))
```

```python
sigmoid=SVC(kernel='sigmoid')
sigmoid.fit(x_train,y_train)
## Prediction
y_predicted4=sigmoid.predict(x_test)
print(classification_report(y_test,y_predicted4))
print(confusion_matrix(y_test,y_predicted4))
```

```python
from sklearn.model_selection import GridSearchCV
 
# defining parameter range
param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['linear']}
```

```python
grid=GridSearchCV(SVC(), param_grid=param_grid, cv=5,verbose=3, refit=True)
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_predicted5=grid.predict(x_test)
print(classification_report(y_test,y_predicted5))
print(confusion_matrix(y_test,y_predicted5))
```
