```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
dataset=pd.read_csv('Algerian_forest_fires_cleaned_dataset.csv', header=1)
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
### starting the data cleaning

### check the missing values

dataset[dataset.isnull().any(axis=1)]
```

```python
dataset.loc[:122, 'reigion'] = 0
dataset.loc[122:, 'reigion'] = 1
```

```python
dataset
dataset.info()
```

```python
dataset[['reigion']]=dataset[['reigion']].astype(int)
```

```python
dataset
```

```python
dataset.isnull().sum()
```

```python
df=dataset.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[121]]
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
### fix the spaces in column names

df.columns
```

```python
df.columns=df.columns.str.strip()
```

```python
df.columns
```

```python
df.info()
```

```python
### change the required columns as integer datatype

df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

```python
### changing the other columns to float datatype

objects = [features for features in df.columns if df[features].dtypes == 'O']
```

```python
objects
```

```python
for i in objects:
    if i!= 'Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
### savong the clean dataset

df.to_csv('Algerian_Dataset_Worked', index=False)
```

```python
### exploratory data analysis
### drop day month and year
df_to_work = df.drop(['day', 'month', 'year'], axis=1)
```

```python
df_to_work.head()
```

```python
df_to_work.to_csv("Algerian_Forest_Dataset_with_only_needed_cols", index=False)
```
