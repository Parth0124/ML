```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
iris
```

```python
print(iris['DESCR'])
```

```python
### create independent and dependent features.
x=pd.DataFrame(iris['data'], columns=['sepal length in cm','sepal width in cm','petal length in cm','petal width in cm'])
```

```python
y=iris['target']
```

```python
y
```

```python
### train test split

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20, random_state=10)
```

```python
### apply decision tree classifier
### postpruning
from sklearn.tree import DecisionTreeClassifier
treeclassifier = DecisionTreeClassifier(max_depth=2)
```

```python
treeclassifier.fit(x_train,y_train)
```

```python
### visualize the decision tree
```

```python
from sklearn import tree
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier,filled=True)
```

```python
y_predict=treeclassifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
accuracy=accuracy_score(y_test,y_predict)
matrix=confusion_matrix(y_test,y_predict)
report = classification_report(y_test,y_predict)
print(accuracy)
print(matrix)
print(report)
```

```python
### prepruning
```

```python
params = {
    'criterion':['gini', 'entropy', 'log_loss'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto', 'sqrt', 'log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid=GridSearchCV(treeclassifier,param_grid=params,cv=5,scoring='accuracy')
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
grid_predicted_y=grid.predict(x_test)
```

```python
accuracy_preprunning = accuracy_score(y_test,grid_predicted_y)
matrix_preprunning=confusion_matrix(y_test,grid_predicted_y)
report_prunning = classification_report(y_test,grid_predicted_y)
print(accuracy_preprunning)
print(matrix_preprunning)
print(report_prunning)
```
