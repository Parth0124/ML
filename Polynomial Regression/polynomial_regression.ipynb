```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
###  quadratic equation

x = 6 * np.random.rand(100,1) - 3
y = 0.5 * x**2 + 1.5*x + 2 + np.random.randn(100,1)
```

```python
### plotting the above equation

plt.scatter(x,y,color='green')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Dataset for polynomial regression')
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
regression_1 = LinearRegression()
```

```python
x_train
```

```python
regression_1.fit(x_train, y_train)
```

```python
from sklearn.metrics import r2_score

score = r2_score(y_test, regression_1.predict(x_test))
print(score)
```

```python
###visualizing the best fit line

plt.plot(x_train, regression_1.predict(x_train), color='red')
plt.scatter(x_train, y_train)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('polynomial regression')
```

```python
### the above line was the best fit line for the quadratic equation previously declared. now the error was quite high along with the accuracy which was around 79%. The above best fit line is made using linear regression. Below is the ploynomial regresssion for the same dataset. 
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
polynomial = PolynomialFeatures(degree=2, include_bias=True)
x_train_polynomial = polynomial.fit_transform(x_train)
x_test_polynomial = polynomial.transform(x_test)
```

```python
x_train_polynomial
```

```python
x_test_polynomial
```

```python
from sklearn.metrics import r2_score
```

```python
regression = LinearRegression()
regression.fit(x_train_polynomial, y_train)
y_prediction = regression.predict(x_test_polynomial)
score = r2_score(y_test, y_prediction)
print(score)
```

```python
print(regression.coef_)
```

```python
plt.scatter(x_train, regression.predict(x_train_polynomial))
plt.scatter(x_train, y_train)
```

```python
from sklearn.pipeline import Pipeline
```

```python
### using pipelining for the above polynomial regression for various degrees.

def poly_regression(degree):
    x_new = np.linspace(-3, 3, 200).reshape(200, 1)
    
    poly_features=PolynomialFeatures(degree=degree,include_bias=True)
    lin_reg=LinearRegression()
    poly_regression=Pipeline([
        ("poly_features",poly_features),
        ("lin_reg",lin_reg)
    ])
    poly_regression.fit(x_train,y_train) ## ploynomial and fit of linear reression
    y_pred_new=poly_regression.predict(x_new)
    #plotting prediction line
    plt.plot(x_new, y_pred_new,'r', label="Degree " + str(degree), linewidth=2)
    plt.plot(x_train, y_train, "b.", linewidth=3)
    plt.plot(x_test, y_test, "g.", linewidth=3)
    plt.legend(loc="upper left")
    plt.xlabel("X")
    plt.ylabel("y")
    plt.axis([-4,4, 0, 10])
    plt.show()
    
```

```python
poly_regression(15)
```
